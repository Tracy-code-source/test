```python
X = Boston_data.iloc[:, 0: 13].values
X = np.insert(X, 0, values = 1, axis = 1)
y = Boston_data.iloc[:, -1].values
w = np.zeros(X.shape[1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 888)
alpha = 0.000001
ite = 1000000
```

```python
def gradient_descent(X, y, w, alpha, ite):
    cost_remain = [] # 记录损失均方差
    num = X.shape[0]
    for i in range (0, ite, 1):
        Loss = np.dot(X, w) - y
        cost = np.sum(Loss ** 2) / (num * 2)
        cost_remain.append(cost)
        dw = (1 / num) * np.dot(X.T, Loss)
        w -= alpha * dw

    return w, cost_remain
```

```python
final_w, remain = gradient_descent(X_train, y_train, w, alpha, ite)
```

```python
y_test_forecast = np.dot(X_test, final_w)
plt.scatter(X_test[:, 6], y_test)
plt.scatter(X_test[:, 6], y_test_forecast)
```
