```python
# 导入相应库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 导入数据集
Boston_data = pd.read_csv(r'C:\Users\HUANG\Downloads\house_data.csv')
```

```python
Boston_data.shape
```

```python
Boston_data.head()
```

```python
def Func():
    num = Boston_data.shape[1] - 1 # 特征数不包括房价中位数
    X = Boston_data.iloc[:, 0: num].values # 读取非房价数据（特征）
    y = Boston_data.iloc[:, -1].values # 房价，即目标向量
    return X, y
```

```python
X, y = Func()
X = np.array(X)
y = np.array(y)
# 为进行矩阵运算，插入偏置项
X = np.insert(X, 0, values = 1, axis = 1)
```

```python
# 设置参数w
w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
```

```python
forecast = np.dot(X, w)
```

```python
# 通过前面工作的可视化，了解到RM与房价存在较强的相关性
```

```python
X_rm = Boston_data.iloc[:, 6].values
```

```python
plt.scatter(X_rm, y)
plt.plot(X_rm, forecast)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
clf.fit(X_rm.reshape(-1, 1), y)
y_pre = clf.predict(X_rm.reshape(-1, 1))
```

```python
plt.scatter(X_rm, y)
plt.plot(X_rm, y_pre)
```
