```python
# 导入相应库
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 尝试用梯度下降法求解 f(x) = x^2 最小值
```

```python
def Build():
    x = np.linspace(-4, 4, 100) # 生成-4到4一百个等间距数值，并赋值给x
    y = x ** 2
    plt.plot(x, y)
```

```python
Build() #调用函数显示图像
```

```python
# 因为是一元，梯度直接对其求导
# 得到导数2x
```

```python
# 初始化迭代起点x,alaph,迭代次数
x = 3
alaph = 0.2
it = 200
```

```python
for i in range(0,it, 1):
    x = x - alaph * (2 * x)
```
